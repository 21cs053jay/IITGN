```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
i = 1
for file in os.listdir('./csvfiles'):
    df = pd.read_csv(f'./csvfiles/{file}')
    df = df[3:]
    if i==10: break
    print(file)
    plt.figure()
    turn = 1
    for x, y in zip(df.iloc[:,1], df.iloc[:,6]):
        if turn==1:

            end_point = (x, 0.0225)
            # print(end_point)
        
            len = round((0.0225-(y/2)),4)
            start_point = (x, len)
            # print(start_point)
            plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], linewidth=1, color='g')

            len = round((0.0225 + (y/2)),4)
            end_point = (x, len)
            # print(end_point)

            start_point = (x, 0.0225)
            # print(start_point)

            plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], linewidth=1, color='g')
            turn = 2

        elif turn==2:
            end_point = (x, 0.0375)
            # print(end_point)

            len = round((0.0375-(y/2)),4)
            start_point = (x, len)
            # print(start_point)

            plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]],linewidth=1, color='b')

            len = round((0.0375 + (y/2)),4)
            end_point = (x, len)
            # print(end_point)


            start_point = (x, 0.0375)
            # print(start_point)
            plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], linewidth=1, color='b')
            turn = 3

        else:
            end_point = (x, 0.0525)
            # print(end_point)
            len = round((0.0525-(y/2)),4)
            start_point = (x, len)
            # print(start_point)
            plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], linewidth=2, color='k')

            len = round((0.0525 + (y/2)),4)
            end_point = (x, len)
            # print(end_point)

            start_point = (x, 0.0525)
            # print(start_point)

            plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]], linewidth=2, color='k')
            turn = 1

    plt.ylim(0.01,.06)
    plt.show()
    i+=1
```

```python
#Just last time intervals 
i = 1
final_5_df = pd.DataFrame()
for file in os.listdir('./csvfiles'):
    if i%5==0:
        df = pd.read_csv(f'./csvfiles/{file}')
        df = df[21:-21]
        split = file.split('_')
        split[2] = split[2].split('.')[0]
        split[1] = float(split[1])
        df['HCl_flux'] = split[1]
        df['time'] = int(split[2])
        final_5_df = pd.concat([final_5_df,df], ignore_index=True)
    i+=1
```

```python
#Just last time intervals - Just avg
i = 1
avg_df = pd.DataFrame()
for file in os.listdir('./csvfiles'):
    if i%5==0:
        df = pd.read_csv(f'./csvfiles/{file}')
        df = df[21:-21]
        split = file.split('_')
        split[2] = split[2].split('.')[0]
        split[1] = float(split[1])
        avg = df.mean()
        dict = {'HCl_flux':split[1],'time':int(split[2]), 'Average Roughness':0.005 - avg[6]}
        avg_df = avg_df._append(dict, ignore_index=True)
    i+=1
```
